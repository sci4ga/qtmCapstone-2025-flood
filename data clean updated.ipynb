```python

def load_storm_data():
    folder_path = "/Users/paranxiaair/Desktop/STUDY/EMORY/大三下/QTM Capstone/Deforestration-and-Flooding-in-GA/Storm Data 1985-2024"  
    pattern = os.path.join(folder_path, "StormDetail_*.csv")  
    
    csv_files = glob.glob(pattern)  # Returns a list of all matching CSV file paths
    
    dataframes = []
    for file_path in csv_files:
        df = pd.read_csv(file_path)
        dataframes.append(df)
    
    if dataframes:
        combined_df = pd.concat(dataframes, ignore_index=True)
        return combined_df
    else:
        return pd.DataFrame()

storm_df = load_storm_data()
storm_df
```

```python

def load_location_data(folder_path):
    """
    1) Loads all 'StormEvents_locations-ftp_v1.0_dXXXX_*.csv' files from 'folder_path'.
    2) Concatenates them into a single DataFrame.
    3) Renames each file from 'StormEvents_locations-ftp_v1.0_dYYYY_c...csv'
       to 'Location_YYYY.csv'.
    """

    # Construct the pattern to find files
    pattern = os.path.join(folder_path, "*.csv")
    csv_files = glob.glob(pattern)

    # Load them into a list of DataFrames
    dataframes = []
    for file_path in csv_files:
        location = pd.read_csv(file_path)
        dataframes.append(location)

    # Combine all DataFrames into one
    if dataframes:
        combined_location = pd.concat(dataframes, ignore_index=True)
        return combined_location
    else:
        return pd.DataFrame()


folder = "/Users/paranxiaair/Desktop/STUDY/EMORY/大三下/QTM Capstone/Deforestration-and-Flooding-in-GA/Location 1985-2024"
location_df = load_location_data(folder)
location_df
```

```python
def generate_map(merged_details, beginyearmonth):
    # Filter the dataset for the given BEGIN_YEARMONTH
    month_data = merged_details[merged_details['BEGIN_YEARMONTH'] == beginyearmonth]
    
    # Extract latitudes and longitudes
    latitudes = month_data['LATITUDE']
    longitudes = month_data['LONGITUDE']
    
    # Create a folium map centered on the first location
    m = folium.Map(location=[latitudes.iloc[0], longitudes.iloc[0]], zoom_start=10)
    
    # Add markers for each location in the dataset
    for lat, lon in zip(latitudes, longitudes):
        folium.Marker([lat, lon]).add_to(m)
    
    # Add a title to the map using Leaflet Control
    month_title = f"Map for {str(beginyearmonth)[:4]}-{str(beginyearmonth)[4:]}"  # Title based on the month
    m.get_root().html.add_child(folium.Element(f'''
        <div style="position: relative; top: 10px; left: 50%; transform: translateX(-50%); font-size: 20px; font-weight: bold; color: black; background-color: white; padding: 10px; border-radius: 5px; box-shadow: 2px 2px 5px rgba(0, 0, 0, 0.3);">{month_title}</div>
    '''))
    
    # Save the map as an HTML file to view interactively
    m.save('map.html')
    
    # Return the map object
    return m
```

```python
beginyearmonth = 202302  
map_object = generate_map(merged_details, beginyearmonth)
map_object
```
